## Months with greater demands

```python
calendar_2['date'] = pd.to_datetime(calendar_2['date'])
calendar_2['price'] = calendar_2['price'].str.replace(',', '')
calendar_2['price'] = calendar_2['price'].str.replace('$', '')
calendar_2['price'] = calendar_2['price'].astype(float)
calendar_2['date'] = pd.to_datetime(calendar_2['date'])
mean_of_month = calendar_2.groupby(calendar_2['date'].dt.strftime('%B'),
                                 sort=False)['price'].mean()
mean_of_month.plot(kind = 'barh' , figsize = (12,7))
plt.xlabel('average monthly price');
```

### The Airbnb price in NC increases in the months of July, October and November.

```python
calendar_2['dayofweek'] = calendar_2.date.dt.weekday_name
cats = [ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
price_week=calendar_2[['dayofweek','price']]
price_week = calendar_2.groupby(['dayofweek']).mean().reindex(cats)
price_week.drop('listing_id', axis=1, inplace=True)
price_week.plot()
ticks = list(range(0, 7, 1)) # points on the x axis where you want the label to appear
labels = "Mon Tues Weds Thurs Fri Sat Sun".split()

plt.xticks(ticks, labels);
```

```python
columns = ['price', 'availability_365', 'calculated_host_listings_count', 'number_of_reviews']
listings_3 = listings_2 [columns]
scaler = MinMaxScaler()
listing_Xfrm = listings_2.copy()
listing_Xfrm [columns] = scaler.fit_transform(listings_3)
```

```python
listing_Xfrm.head()
```

```python
linear_model = smf.ols(formula='price ~ C(room_type) + C(neighbourhood) + availability_365 + calculated_host_listings_count + number_of_reviews', data=listing_Xfrm)
fitted_model = linear_model.fit()
fitted_model.summary()
```
